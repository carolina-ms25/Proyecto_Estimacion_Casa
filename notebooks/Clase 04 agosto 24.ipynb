```python
# Aprendizaje Maquina

# Data
data_dict = {"base":[2, 4, 6, 8],"altura":[4, 2, 4, 2],"area":[4, 4, 12, 8]}
import pandas as pd
data = pd.DataFrame(data_dict)
data

# Aplicar Algoritmo ML a Data
from sklearn.linear_model import LinearRegression # Regresion lineal
x = data[["base","altura"]]
y = data["area"]
modelo = LinearRegression()
modelo.fit(x,y)
print(f"B0= {modelo.intercept_}")
print(f"B1, B2 ={modelo.coef_}")

# Estimar area utilizando ML
base, altura = 2, 4
area = modelo.intercept_ + modelo.coef_[0]*base + modelo.coef_[1]*altura
area
```
